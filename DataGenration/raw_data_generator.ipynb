```python
import boto3
import os
import pandas as pd
import random
from faker import Faker
import datetime
import time
```

```python
faker=Faker()
date=faker.date_time()
timestamp=time.mktime(date.timetuple())
['Asia','Africa','Europe','North America','South America','Australia','Antarctica']
Actives = []
for _ in range(100):
    x=faker.password(length= 8, special_chars=False)
    a=faker.city()
    b=faker.country()
    c=faker.address()
    y=random.choice(['Asia','Africa','Europe','North America','South America','Australia','Antarctica'])
    d=faker.state()
    date=faker.date_time()
    e=time.mktime(date.timetuple())
    f=faker.swift(8)
    g=faker.latitude()
    h=faker.longitude()
    j=faker.date()
    Actives.append([x,a,b,y,[c],d,int(e),f,g,h,int(j.split("-")[1]),j])
print(Actives)
```

```python
faker=Faker()
date=faker.date_time()
timestamp=time.mktime(date.timetuple())
channels=open(r"C:\Users\nitesh.singh\Downloads\channel.txt")
channel_list=channels.read().split(",")
viewership=[]
for _ in range(100):
    a=faker.password(length= 8, special_chars=False)
    b=random.choice(['News','Spiritual','Music','Sports','Entertainment','Cartoon','Movies','Business','Infotainment','Comedy'])
    c=random.choice(channel_list)
    d=faker.city()
    e=random.choice(["Laptop","TV","Tablet","Mobile"])
    f=random.choice(["Window","Andriod","Cabel"])
    g=random.randint(900,18001)
    h=faker.password(6,False,True,False)
    i=faker.language_name()
    j=faker.country()
    k=random.choice(['Asia','Africa','Europe','North America','South America','Australia','Antarctica'])
    l=faker.address()
    date=faker.date_time()
    m=time.mktime(date.timetuple())
    n=random.choice(['News','Spiritual','Music','Sports','Entertainment','Cartoon','Movies','Business','Infotainment','Comedy'])
    o=faker.name()
    p=faker.state()
    q=faker.latitude()
    r=faker.longitude()
    t=faker.date()
    viewership.append([a,b,c,d,e,f,g,h,i,j,k,[l],int(m),n,o,p,q,r,t.split("-")[1],t])
print(viewership)
```
